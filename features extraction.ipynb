```python
import scipy.io
import matplotlib.pyplot as plt

# Load the .mat file
data = scipy.io.loadmat(r"D:\NITJ\SEED\Preprocessed_EEG\1_20131027.mat")

# Choose trial 1
eeg = data['djc_eeg1']  # shape: (62, 47001)

# Set time axis (assuming 200 Hz sampling rate)
sampling_rate = 200  # Hz
time = [i / sampling_rate for i in range(eeg.shape[1])]  # in seconds

# Plot 5 channels (e.g., channels 0 to 4)
plt.figure(figsize=(12, 8))

for i in range(1):
    plt.plot(time, eeg[i], label=f'Channel {i+1}')

plt.title("EEG Signal - Trial 1 (Channels 1)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (uV)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import scipy.io as sio
import os

eeg_folder = r"D:\Preprocessed_EEG"
files = sorted(f for f in os.listdir(eeg_folder) if f.endswith('.mat'))

for fn in files:
    m = sio.loadmat(os.path.join(eeg_folder, fn))
    k0 = [k for k in m.keys() if not k.startswith('__')][0]
    print(fn, m[k0].shape)
```
